```python
import json
from collections import defaultdict
```

```python

with open("brasil_municipios_Geo.geojson", "r", encoding="utf-8") as f:
    geo = json.load(f)
```

```python
por_uf = defaultdict(list)

for feature in geo["features"]:
    uf = feature["properties"]["id"][:2]
    por_uf[uf].append(feature)

for uf, features in por_uf.items():
    novo_geo = {
        "type": "FeatureCollection",
        "features": features
    }
    with open(f"{uf}.json", "w", encoding="utf-8") as f:
        json.dump(novo_geo, f, ensure_ascii=False, indent=2)
```
