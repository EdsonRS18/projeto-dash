```python
import pandas as pd
import numpy as np
from pathlib import Path

# -----------------------------
# Caminho do arquivo original
# -----------------------------
input_file = "df_geografico_finalizado2.csv"

# pasta onde serão salvas as partes
output_dir = Path("data_parts")
output_dir.mkdir(exist_ok=True)

# -----------------------------
# Carregar dataframe
# -----------------------------
df = pd.read_csv(input_file)

# -----------------------------
# Dividir em 4 partes
# -----------------------------
partes = np.array_split(df, 4)

# -----------------------------
# Salvar partes
# -----------------------------
for i, parte in enumerate(partes, start=1):
    file_name = output_dir / f"df_parte_{i}.csv"
    parte.to_csv(file_name, index=False)
    print(f"Salvo: {file_name} | Linhas: {len(parte)}")

print("Divisão concluída.")
```
